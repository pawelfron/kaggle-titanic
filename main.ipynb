```python
from sklearn.preprocessing import OrdinalEncoder
def transform(dataset: pd.DataFrame):
    X = dataset[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']]

    ordinal_enc = OrdinalEncoder()

    X = pd.get_dummies(X, columns=['Sex', 'Embarked'], drop_first=True)
    X['Cabin'] = ordinal_enc.fit_transform(X[['Cabin']])

    return X
```

```python
test = pd.read_csv('data/test.csv')
test.describe(include='all')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

X_train = transform(train)
y_train = train['Survived']
X_test = transform(test)

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier())
])
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
```

```python
import os
os.makedirs('output', exist_ok=True)

result = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': y_pred})
result.to_csv('output/result.csv', index=False)
```
